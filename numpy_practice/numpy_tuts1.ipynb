## numpy indexing and slicing of vectors

```python
import numpy as np
```

```python
arr = np.arange(1, 11)
arr
```

```python
arr[9] # access the 10th element (index starts from 0)
```

```python
arr[1:5] # slice the array from index 1 to 4 (5 is exclusive)
```

```python
arr[:5] # slice the array from the beginning to index 4 (5 is exclusive)
```

```python
arr[3:] # slice the array from index 3 to the end
```

```python
arr[3::2] # slice the array from index 3 to the end, with a step of 2 (every other element)
```

## numpy indexing and slicing of matrix

```python
arr = np.arange(1, 31).reshape(5, 6)
arr
```

```python
arr[0] # first row
```

```python
arr[3] # fourth row (index starts from 0)
```

```python
arr[0,0] # first element of the first row
```

```python
arr[0,5] # last element of the first row
```

```python
arr[4,5] # last  element of the last row
```

```python
# arr[start_row:end_row, start_col:end_col]
arr[0:2, 0:3] # slice the first two rows and the first three columns
```

```python
arr[0:2, 1:3] # slice the first two rows and the second and third columns
```

```python
arr
```

```python
arr[3:5, 3:6] # slice the last two rows and the last three columns
```

```python
arr[:5] # slice the array from the beginning to index 4 (5 is exclusive)
```

```python
arr[:,2] # slice all rows and the third column (index 2)
```

## Boolean Indexing

```python
arr = np.arange(12, 21)
arr
```

```python
bool_index = arr%2 == 0 # create a boolean index for even numbers
bool_index
```

```python
arr = arr[bool_index] # use the boolean index to filter the array
arr
```

## Array Operations

### Arithmetic operations

```python
a1 = np.array([1,2,3,4,5])
a2 = np.array([6,7,8,9,10])
```

```python
# element size must be the same for array operations
a1 + a2 # element-wise addition
```

```python
a1-a2 # element-wise subtraction
```

```python
a1*a2 # element-wise multiplication
```

```python
a1/a2 # element-wise division
```

```python
a1//a2 # element-wise floor division (integer division)
```

```python
a1**2 # element-wise exponentiation (squaring each element)
```

```python
l = [10, 20, 30, 40, 50]
arr = np.array(l) # convert list to numpy array
```

```python
arr+10 # add 10 to each element of the array
```

```python
arr2 = np.arange(1,26).reshape(5,5)
arr2
```

```python
arr2 = arr2 + 10 # add 10 to each element of the array
arr2
```

```python
# Deep copy vs Shallow copy
a = np.arange(1, 21)
a
```

```python
slice = a[0:5] # create a slice of the array (shallow copy)
slice = slice + 10 # modify the slice
slice
```

```python
a
```

```python
a1 = a[4:6]**2 # modify the shallow copy (slice) and assign it to a new variable (a1)
a1 
```

```python
a
```

## Matrix Operations

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
```

```python
A
```

```python
B
```

```python
# multiply A and B element-wise
A * B
```

```python
A @ B # matrix multiplication (dot product) of A and B
```

```python
np.dot(A, B) # another way to perform matrix multiplication (dot product) of A and B
```

```python
# transpose of A
A.T
```

```python
A
```

```python
B
```

```python
# stck array
a = np.arange(1, 11).reshape(2, 5)
b = np.arange(11, 21).reshape(2, 5)
```

```python
a
```

```python
b
```

```python
# vertical stack (stack arrays vertically)
np.vstack((a, b))
```

```python
# horizontal stack (stack arrays horizontally)
np.hstack((a, b))
```

```python
# spliting arrays
arr = np.arange(16).reshape(4, 4)
arr
```

```python
# horizontal split the array into 2 sub-arrays
np.hsplit(arr, 2)
```

```python
np.hsplit(arr, 3) ## show error: array split does not result in an equal division
```

```python
np.split(arr, 2)
```
