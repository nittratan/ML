## Reshaping and resizing

```python
arr = np.arange(1, 31) # creates a 1D array with values from 1 to 30
arr
```

```python
arr.reshape(5, 6) # reshapes the array to 5 rows and 6 columns
# if the total number of elements does not match, it will raise an error
# 5*6 = 30, which matches the total number of elements in the original array (1 to 30)
```
