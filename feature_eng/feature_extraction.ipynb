### create a new column for episode count

```python
def extract_episode(txt):
    check = False
    data = ''
    
    for i in txt:
        if i == ')':
            break
        
        if check:
            data += i
        
        if i == '(':
            check = True
            
    return data
```

```python
df['Title'].apply(extract_episode) # apply the extract_episode function to each value in the Title column
```

```python
df['Episodes'] = df['Title'].apply(extract_episode) # create a new column for episode count
```

### Create a new column for time stamps

```python
df.loc[0]['Title'] # return the title of the first anime
```

```python
import re

def extract_timestamp(txt):
    match = re.search(r'\)([A-Z][a-z]{2} \d{4} - [A-Z][a-z]{2} \d{4})', txt)
    
    if match:
        return match.group(1)
    return None
```

```python
df['Timestamp'] = df['Title'].apply(extract_timestamp) # create a new column for timestamp
```
