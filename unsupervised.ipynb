```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=0.60, random_state=42)
```

```python
X
```

```python
df = pd.DataFrame(X, columns=['Feature 1', 'Feature 2'])    
df.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
intertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    intertia.append(kmeans.inertia_)
```

```python
intertia
```

```python
plt.plot(range(1, 11), intertia, marker='o')
```

```python
kmeans_final = KMeans(n_clusters=3, random_state=42)
```

```python
cluster_labels = kmeans_final.fit_predict(X_scaled)
```

```python
cluster_labels
```

```python
df['Cluster'] = cluster_labels
sns.scatterplot(data=df, x='Feature 1', y='Feature 2', hue='Cluster', palette='Set1')
plt.title('K-Means Clustering Results')
plt.show()
```

```python
from sklearn.datasets import make_moons
```

```python
X, y_true = make_moons(n_samples=300, noise=0.5, random_state=42)
```

```python
from sklearn.cluster import DBSCAN, KMeans
```

```python
df = pd.DataFrame(X, columns=['Feature 1', 'Feature 2'])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)
```

```python
df['kmeans_cluster'] = kmeans_labels
```

```python
sns.scatterplot(data=df, x='Feature 1', y='Feature 2')
plt.title('K-Means Clustering on Moons Dataset')
plt.show()
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)
```

```python
df['dbscan_cluster'] = dbscan_labels
```

```python
sns.scatterplot(x = df['Feature 1'], y = df['Feature 2'], hue=df['dbscan_cluster'], palette='Set1')
plt.title('DBSCAN Clustering on Moons Dataset')
plt.show()
```
