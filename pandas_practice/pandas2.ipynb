## Pandas Dataframe

##### A Pandas DataFrame is a two-dimensional labeled data structure in Pandas. It is similar to a table in a database or an Excel spreadsheet, with data arranged in rows and columns.

##### Each column in a DataFrame can have a different data type, such as integers, floats, strings, or dates.

```python
import numpy as np
import pandas as pd
```

# Creating a Dataframe

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
    'salary': [50000, 60000, 70000, 80000]
}
```

```python
data
```

```python
pd.DataFrame(data)
```

```python
df = pd.DataFrame(data)
```

```python
df
```

### creating a new Column

```python
df2['Designation'] = ['Engineer', 'Manager', 'Director', 'VP'] # add a new column to df2
df2
```

### Remove column
~~~
axis=0       -> rows
axis=1       -> columns
inplace=True -> change original DataFrame
~~~


```python
# df2.drop('Designation') # throw error because Designation is not an index, it's a column
```

```python
df2.drop('Designation', axis=1) # drop the 'Designation' column from df2 
# not drop from original df2, just return a new dataframe without the 'Designation' column
```

```python
df2.drop('Designation', axis=1, inplace=True) # drop the 'Designation' column from 
# df2 and modify df2 in place
df2
```

### Selecting a row

```python
df.loc[0] # access the first row of df2 using .loc
```

```python
df
```

```python
# selecting subset of rows and columns using .loc
df.loc[0:2, ['Name', 'City']] # select rows 0 to 2 and columns 'Name' and 'City' using .loc
```

```python
df.iloc[3] # access the fourth row of df2 using .iloc
```

```python
df.loc[[0,1]][['Name', 'City']] # select rows 0 and 1 and columns 'Name' and 'City' using .loc
```

### conditional Selection

```python
# view only those people who are older than 30
df[df['Age'] > 30] # inner condition to filter rows where 'Age' is greater than 30
```

```python
# those people who are older than 30 and live in Chicago
df[(df['Age'] > 30) & (df['City'] == 'Chicago')] # inner conditions to filter rows where 'Age' is greater than 30 and 'City' is 'Chicago'
```
