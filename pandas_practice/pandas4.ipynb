## Pandas - merge, join, and concat 

```python
import numpy as np
import pandas as pd
```

```python
# employees DataFrame with employee_id, Name, and Department columns
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Department': ['HR', 'IT', 'IT', 'HR', 'Finance'],
    
})

# salaries DataFrame with employee_id and Salary columns
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5, 6],  # Note that employee_id 6 does not exist in the employees DataFrame
    'Salary': [70000, 80000, 75000, 72000, 90000, 85000],
    'bonus': [5000, 6000, 5500, 5200, 7000, 6500]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees, salaries)
```

```python
# Merge the employees and salaries DataFrames on employee_id
merged_data = pd.merge(employees, salaries, on='employee_id')
```

```python
merged_data
```

```python
pd.merge(employees, salaries, on='employee_id', how='inner')  # Default is 'inner', so this will give the same result as above
```

```python
pd.merge(employees, salaries, on='employee_id', how='outer')  # This will include all records from both DataFrames, filling in NaN for missing values
```

```python
pd.merge(employees, salaries, on='employee_id', how='left')
```

```python
pd.merge(employees, salaries, on='employee_id', how='right')
```

## Concatination of 2 dataframes

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})
df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df2, df1])
```

```python
pd.concat([df1, df2], ignore_index=True)
```

## Joining 2 data frames

```python
df1 = pd.DataFrame({
    'name' : ['Alice', 'Bob', 'Charlie'],
}, index=[1, 2, 3])

df2 = pd.DataFrame({
    'score' : [85, 90, 95],
}, index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2, how='outer')
```

```python
df1.join(df2, how='inner')
```

```python
df2.join(df1, how='inner')
```

```python
df2.join(df1)
```
