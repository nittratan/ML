## Missing Data
fillna() on a selected column.
dropna() is row-wise by default.

```python
import numpy as np
import pandas as pd
```

### Finding missing data

```python
data = {
    'A': [1,2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, 8, 10],
    'C': [10, 20, 30, 40, 50],
    'D': [np.nan, 1, 2, 3, 4]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna() # check for missing values in the DataFrame, returns a DataFrame of the same shape with True for missing values and False for non-missing values
```

```python
df.isna().sum() # count the number of missing values in each column, returns a Series with the count of missing values for each column
```

```python
df.isna().any() # check if there are any missing values in each column, returns a Series with True for columns that have at least one missing value and False for columns that have no missing values
```

```python
df.dropna() # drop rows with any missing values, returns a new DataFrame with only the rows that have no missing values
```

```python
df
```

```python
df.dropna(thresh=3) # drop rows that have less than 3 non-missing values, returns a new DataFrame with only the rows that have at least 3 non-missing values
```

### Filling the missing Data

```python
df
```

```python
df.fillna(0) # fill missing values with 0, returns a new DataFrame with missing values replaced by 0
# not apply to original DataFrame unless you use inplace=True
```

```python
values = {'A': 0, 'B': df['B'].mean(), 'D': df['D'].median()} # create a dictionary with column names as keys and fill values as values
df.fillna(values) # fill missing values with specified values for each column, returns a new DataFrame with missing values replaced by the specified values for each column
```
