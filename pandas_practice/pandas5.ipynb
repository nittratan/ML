## Group by Aggregation

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'category': ['A', 'B', 'A', 'B', 'A','B', 'A', 'B'],
    'store': ['S1', 'S2', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'sales': [100, 150, 200, 250, 300, 350, 400, 450],
    'quantity': [10, 15, 20, 25, 30, 35, 40, 45],
    'date': pd.date_range(start='2023-01-01', periods=8, freq='D')
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# group by category and calculate total sales
cat =df.groupby('category')
cat
```

```python
for name, group in cat:
    print(f"Category: {name}")
    print(group)
    print("\n")
```

```python
cat = df.groupby('category')['sales'].sum()
```

```python
cat
```

```python
df.groupby(['category', 'store'])['sales'].sum()
```

```python
# Aggregation functions
# mean, sum, count, min, max, std, var, median
df.groupby('category')['sales'].agg(['mean', 'sum', 'count', 'min', 'max', 'std', 'var', 'median'])
```
