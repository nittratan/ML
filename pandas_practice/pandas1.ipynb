# Pandas Series

A Pandas Series is a one-dimensional labeled array that can hold data of any type, such as integers, floats, strings, or Python objects. It is similar to a column in a table or spreadsheet. Each value in a Series has an index, which can be numeric or custom-labeled.

### A NumPy array is like a simple list of values, while a Pandas Series is like a NumPy array with labels/index attached to each value.

#### numpy array
import numpy as np
arr = np.array([10, 20, 30])
print(arr)
output [10 20 30]

#### Pandas Series
import pandas as pd

s = pd.Series([10, 20, 30], index=["a", "b", "c"])
print(s)
output
a    10
b    20
c    30
dtype: int64

```python
import numpy as np
import pandas as pd
```

```python
lebels = ['a', 'b', 'c']
my_data = [10, 20, 30]
arr = np.array(my_data)
d = {'a': 10, 'b': 20, 'c': 30}
pd.Series(data=my_data)
```

```python
pd.Series(data=my_data, index=lebels) # create a series from a list, the index will be the default integer index and the data will be the values in the list
```

```python
pd.Series(arr) # create a series from a numpy array, the index will be the default integer index and the data will be the values in the array
```

```python
pd.Series(d) # create a series from a dictionary, the keys become the index and the values become the data
```
