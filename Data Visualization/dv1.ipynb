### DPI in Matplotlib

DPI stands for **dots per inch**. In Matplotlib, it controls the resolution or sharpness of a figure.

The final image size in pixels is calculated using:

**Pixels = Figure size in inches × DPI**

For example, if a figure has `figsize=(8, 4)` and `dpi=100`, then the output size will be:

**Width:** 8 × 100 = 800 pixels  
**Height:** 4 × 100 = 400 pixels

So, `figsize` controls the shape and size of the figure, while `dpi` controls the image quality.

```python
# object oriented way
fig = plt.figure() # create a figure
axis1 = fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, width, height
axis1.plot(x,y)
axis2 = fig.add_axes([0.2,0.5,0.4,0.3]) # left, bottom, width, height
axis2.plot(y,x)
```
