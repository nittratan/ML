## Matrix Plot

A **matrix plot** is used to display data in a grid or matrix format. It is useful when we want to visualize relationships between two variables or compare values across rows and columns.

In Seaborn, matrix plots are commonly used with **heatmaps** and **clustermaps**.

---

### 1. Heatmap

A **heatmap** represents values using colors. Higher or lower values are shown with different color intensities.

```python
sns.heatmap(data)
```
### 2. Clustermap
A clustermap is similar to a heatmap, but it also groups similar rows and columns together using clustering.

```python
sns.clustermap(data)
```

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
````

````python
flights = sns.load_dataset('flights')
````

````python
flights.head()
````

````python
tips = sns.load_dataset('tips')
````

````python
tips
````

````python
# heatmap: to visualize the correlation between two variables
# example of correlation matrix
````

````python
tipscorr = tips[['total_bill', 'tip', 'size']].corr()
````

````python
tipscorr
````

````python
sns.heatmap(tipscorr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Tips Dataset')
plt.show()
````

````python
# clustermap: to visualize the correlation between two variables with hierarchical clustering
sns.clustermap(tipscorr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Tips Dataset with Clustering')
plt.show()
````

````python
# pivot table with heatmap
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(flights_pivot, annot=True, cmap='coolwarm')
plt.title('Pivot Table of Flights Dataset')
plt.show()
````
