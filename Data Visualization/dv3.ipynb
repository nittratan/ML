## Categorical Plot

A **categorical plot** is used to visualize data where one variable is made up of categories, such as names, groups, types, or labels.

Categorical plots help us compare values between different categories.

### Common Categorical Plot Functions

#### 1. Bar Plot

A **bar plot** shows values for different categories using rectangular bars.

```python
plt.bar(categories, values)

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
````

````python
df = sns.load_dataset('tips')
df.head()
````

````python
sns.countplot(x = df['sex'], hue = df['smoker'])
````

````python
sns.barplot(x = df['sex'], y = df['total_bill'], hue = df['smoker'], estimator = np.mean)
````

````python
sns.boxplot(data=df, x="day", y="total_bill")
# boxplot: to visualize the distribution of a variable and identify outliers
````

````python
sns.violinplot(x="day", y="total_bill", data=df)
# violinplot: to visualize the distribution of a variable and identify outliers, similar to box
````

````python
# stripplot: to visualize the distribution of a variable and identify outliers, similar to boxplot and violinplot
sns.stripplot(x="day", y="total_bill", data=df, jitter=True)
````
