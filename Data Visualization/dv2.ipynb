```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
df['size'].unique()
```

```python
sns.displot(df['total_bill'], kde=True)
```

```python
sns.displot(df['size'], bins=10, kde=True) # kde is kernel density estimation, it is a way to estimate the probability density function of a random variable. It is a non-parametric way to estimate the probability density function of a random variable. It is a way to smooth the histogram of the data.
```

```python
plt.subplot(1, 2, 1)
sns.displot(df['total_bill'], kde=True)
plt.subplot(1, 2, 2)
sns.displot(df['tip'], bins=10, kde=True)
```

```python
# jointplot: to visualize the relationship between two variables
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter') # kind is the type of plot to draw, it can be 'scatter', 'reg', 'resid', 'kde', 'hex'
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='reg')
```

```python
# pairplot: to visualize the relationship between multiple variables
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

```python
sns.rugplot(df['total_bill'])
```
